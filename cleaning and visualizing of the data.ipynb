## Cleaning the data

### how to handle missing values
### for float datatype we should use or it is best to use median
### for object we should go with mode
### if the missing value is more then drop it


```python
median1= df['LoanAmount'].median()
median1
```

```python
df['LoanAmount'].replace(np.nan,median1,inplace = True)
```

```python
median2 = df['Credit_History'].median()
median2
```

```python
df['Credit_History'].replace(np.nan,median2,inplace = True)
```

```python
df.isnull().sum()
```

```python
median3 = df['Loan_Amount_Term'].median()
median3
```

```python
df['Loan_Amount_Term'].replace(np.nan,median3,inplace = True )
```

```python
df.isnull().sum()
```

```python
mode1 = df['Gender'].mode().values[0]
mode1
df['Gender'].replace(np.nan,mode1,inplace = True)
df.isnull().sum()
```

```python
mode2 = df['Dependents'].mode().values[0]
mode2
df['Dependents'].replace(np.nan,mode2,inplace = True)
mode3 = df['Married'].mode().values[0]
mode3
df['Married'].replace(np.nan,mode3,inplace = True)
mode4 = df['Self_Employed'].mode().values[0]
df['Self_Employed'].replace(np.nan,mode4,inplace = True)
df.isnull().sum()
```

### To remove outliers 

```python
df.boxplot('LoanAmount')
```

```python
def remove_outlier(col):
    sorted(col)
    q1,q3= col.quantile([0.25,0.75])
    iqr = q3-q1
    lower_range=q1-(1.5*iqr)
    upper_range=q3+(1.5*iqr)
    return lower_range,upper_range
lowLoanAmount,uppLoanAmount = remove_outlier(df['LoanAmount'])
df['LoanAmount'] = np.where(df['LoanAmount']>uppLoanAmount,uppLoanAmount,df['LoanAmount'])
df['LoanAmount'] = np.where(df['LoanAmount']<lowLoanAmount,lowLoanAmount,df['LoanAmount'])
```

```python
df.boxplot('LoanAmount')
```
